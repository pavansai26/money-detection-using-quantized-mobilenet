```python
def trainingdata():
    training_data = []
    file='C:/Users/lenovo/Downloads/challenges-in-representation-learning-facial-expression-recognition-challenge/daaata'
    categories=['10 rupees','20 rupees','50 rupees','100 rupees','200 rupees','500 rupees','2000 rupees']
    for category in categories:
        path = os.path.join(file,category)
        class_num = categories.index(category)
        for img in (os.listdir(path)):
            try:
                img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_GRAYSCALE)
                new_array = cv2.resize(img_array, (200,200))
                training_data.append([new_array, class_num])
            except Exception as e:
                pass
    return training_data
```
